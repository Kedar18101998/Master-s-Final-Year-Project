```python
import pandas as pd
import re
from nltk import sent_tokenize


"""
To install nltk:
    * Run pip install nltk
    * Run python -m nltk.downloader popular
"""

list = []
def main():
    # Read clinical_database.xlsx
    df = pd.read_excel('../input/dataset/All canine HxDx.xlsx')

    # Fill blank values in the column 'CaseHistologicalDiagnosis' with 'N/A'
    df = df.fillna(value={'CaseHistologicalDiagnosis': 'N/A'})

    # Use nltk sent_tokenize to breakdown column 'CaseHistologicalDiagnosis' into a list of sentences
    df['CaseHistologicalDiagnosis'] = df['CaseHistologicalDiagnosis'].apply(sent_tokenize)

    itr = 0     # iterator for use as index in the new blank dataframe
    new_df = pd.DataFrame(columns = [   # new blank dataframe, where we will append the processed data
        'CaseNumber', 'CaseBreed', 'CaseAnimalAgeYears', 'CaseAnimalAgeMonths',
        'CaseSex', 'CaseHistologicalDiagnosis', 'Pathology', 'AnatomicalInfectedArea'], dtype=pd.StringDtype())
    
    for index, row in df.iterrows():    # iterate over the original dataframe df
        for sentence in row['CaseHistologicalDiagnosis']:   # iterate over each sentence in the list on column 'CaseHistologicalDiagnosis'
            clean_sentence = ''     # set a variable to store the processed sentence
            if sentence != 'N/A':   # if sentence is not N/A
                dirty = bool(re.match('[A-Z1-9]\W\W|[A-Z1-9]\W', sentence))     # check if there are ordered list symbols (A., B., 1., 2., A & B:, ...), returns True if there are instances found
                if dirty:   # if there are ordered list symbols and other necessary data
                    clean_sentence = str(re.sub('[A-Z1-9]\W\W|[A-Z1-9]\W', '', sentence)).strip()   # replace with blank then strip
                else:   # if there are no unncessary characters in the sentence
                    clean_sentence = sentence   # store the sentence as the processed sentence
            else:       # if sentence is N/A
                clean_sentence = sentence   # store N/A as the processed sentence
            
            if clean_sentence.strip() != '':    # now, if cleaned sentence is not equal to blank when stripped
                # create a list with two values, splitting the cleaned sentence,
                # and using the last instance of the comma as separator
                # this is done using rsplit function
                k = [j.strip().replace('.', '') for j in clean_sentence.replace('-', ',').rsplit(',', 1)] if clean_sentence != 'N/A' else [None, None]
                if len(k) == 1:
                    k.append("")
                # this code adds a row to the blank dataframe for each cleaned sentence
                new_df.loc[itr] = [row['CaseNumber'], row['CaseBreed'], row['CaseAnimalAgeYears'], row['CaseAnimalAgeMonths'], row['CaseSex'], row['CaseHistologicalDiagnosis'], k[0], k[1]]

                # increase index everytime data is added to the dataframe
                itr += 1

    # output the new dataframe to an excel file
    new_df.to_excel('./clinical_database_rewrite.xlsx')

# driver function, ensures that runtime starts with the main() function
if __name__ == '__main__':
    main()
```

```python
import spacy
import scispacy

from scispacy.linking import EntityLinker

nlp = spacy.load("en_core_sci_sm")

# This line takes a while, because we have to download ~1GB of data
# and load a large JSON file (the knowledge base). Be patient!
# Thankfully it should be faster after the first time you use it, because
# the downloads are cached.
# NOTE: The resolve_abbreviations parameter is optional, and requires that
# the AbbreviationDetector pipe has already been added to the pipeline. Adding
# the AbbreviationDetector pipe and setting resolve_abbreviations to True means
# that linking will only be performed on the long form of abbreviations.
nlp.add_pipe("scispacy_linker", config={"resolve_abbreviations": True, "linker_name": "umls"})

doc = nlp(query)

# Let's look at a random entity!
entity = doc.ents[0]

print("Name: ", entity)

# Each entity is linked to UMLS with a score
# (currently just char-3gram matching).
linker = nlp.get_pipe("scispacy_linker")
for umls_ent in entity._.umls_ents:
    Concept_Id, Score = umls_ent
    print('Concept_Id = {} Score = {}'.format(Concept_Id,Score))
    print(linker.umls.cui_to_entity[umls_ent[0]])
```
